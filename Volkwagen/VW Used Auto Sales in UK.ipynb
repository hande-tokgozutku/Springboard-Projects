# Regression

```python
# defining the variables 
x = dc2[['year', 'mileage', 'tax', 'mpg', 'engineSize']]
y = dc2['price']
  
# adding the constant term 
x = sm.add_constant(x) 
  
# performing the regression 
# and fitting the model 
result = sm.OLS(y, x).fit() 
  
# printing the summary table 
print(result.summary())
```

```python

plt.scatter(dc2['price'], result.predict(x))
plt.tight_layout()
plt.title("Regression Plot vs. Actual Prices")
plt.show()
```

## Our multivariate regression model is strong with an R-squared value of 0.756
* The model we built predicts price as an equation of independent variables year, tax, mileage, MPG and engine size.
* The model performs better for prices under 25,000 and skews more when price is higher. 

# Executive Summary of Insights

* The most important factors when determining car price are year, engine size and mileage.
* Non-numeric factors such as transmission, model and fuel type should also be taken into consideration since these also have impact on prices.
* We created a strong OLS Multivariate Regression model which can predict car prices successfully especially when the price is under 25,000 - this corresponds to the vast majority of the prices as seen on the price histogram.
